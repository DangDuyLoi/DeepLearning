```python
print("Cau 13")
class SumModel(nn.Module):
    def forward(self, x, y):
        return x + y
model13 = SumModel()
# Chay thu:
x13 = torch.tensor([5.0])
y13 = torch.tensor([10.0])
print("Ket qua SumModel:", model13(x13, y13))

print("Cau 14")
class DiffProdModel(nn.Module):
    def forward(self, x, y):
        return x - y, x * y
model14 = DiffProdModel()
# Chay thu:
x14 = torch.tensor([8.0])
y14 = torch.tensor([3.0])
hieu, tich = model14(x14, y14)
print(f"Hieu: {hieu.item()}, Tich: {tich.item()}")

print("Cau 15")
class PerimeterModel(nn.Module):
    def forward(self, a, b):
        return (a + b) * 2
model15 = PerimeterModel()
# Chay thu hinh chu nhat dai 4, rong 3:
print("Chu vi hcn:", model15(torch.tensor(4), torch.tensor(3)).item())

print("Cau 16")
class SurfaceAreaModel(nn.Module):
    def forward(self, a, b, c):
        s_xq = (a + b) * c * 2
        s_tp = (a * b) * 2 + s_xq
        return s_tp
model16 = SurfaceAreaModel()
# Chay thu hinh hop co a=4, b=3, c=5:
print("Dien tich toan phan:", model16(torch.tensor(4), torch.tensor(3), torch.tensor(5)).item())

print("Cau 17")
class TrapezoidParallelModel(nn.Module):
    def forward(self, AD, BC, AB, CD):
        return torch.logical_and(AD == BC, AB == CD)
model17 = TrapezoidParallelModel()
# Chay thu voi cac canh bang nhau:
print("Hai canh ben song song:", model17(torch.tensor(5), torch.tensor(5), torch.tensor(4), torch.tensor(4)).item())

print("Cau 18")
class TrapezoidBaseModel(nn.Module):
    def forward(self, AD, BC, is_parallel):
        return torch.logical_and(AD == BC, is_parallel)
model18 = TrapezoidBaseModel()
# Chay thu:
print("Hai canh day bang nhau:", model18(torch.tensor(5), torch.tensor(5), torch.tensor(True)).item())

print("Cau 19")
class MidlineTriangleModel(nn.Module):
    def forward(self, MN, MB, is_MN_parallel_BC):
        return torch.logical_and(MN == MB, is_MN_parallel_BC)
model19 = MidlineTriangleModel()
# Chay thu:
print("NA = NC la:", model19(torch.tensor(3), torch.tensor(3), torch.tensor(True)).item())
```
